## üìå Project Overview
This project demonstrates **Propensity Score Matching (PSM)** to estimate causal effects in observational data where randomized experiments are not feasible.

The notebook implements a complete causal inference workflow:
- Estimating propensity scores
- Matching treated and control units
- Evaluating covariate balance
- Estimating treatment effects

The goal is to reduce selection bias and obtain a more reliable estimate of treatment impact.

---

## üéØ Objectives
- Apply causal inference concepts to real data  
- Model treatment assignment using logistic regression  
- Perform nearest-neighbor matching  
- Evaluate overlap and balance between groups  
- Estimate treatment effects after matching  

---

## üß† Methodology

### 1. Data Preparation
- Define treatment variable  
- Select relevant covariates  
- Clean and preprocess dataset  

### 2. Propensity Score Estimation
- Logistic regression predicts probability of treatment  
- Scores represent likelihood of receiving treatment  

### 3. Common Support Check
- Compare score distributions between groups  
- Ensure overlap exists before matching  

### 4. Matching
- 1:1 nearest neighbor matching  
- Caliper threshold applied  
- Balanced dataset generated  

### 5. Treatment Effect Estimation
- Compare outcomes between matched groups  
- Estimate causal impact
